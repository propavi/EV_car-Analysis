```python
import pandas as pd
import matplotlib.pyplot as plt
ev = pd.read_csv("EVCars.csv")
ev
```

```python
ev_stock = ev[ev["parameter"] == "EV stock"]
ev_stock
```

```python
ev_stock_grouped = ev_stock.groupby("year")["value"].sum()
ev_stock_grouped.plot(x="year",y="value",kind="bar")
plt.xlabel("year")
plt.ylabel("EV Stock")
plt.show()
```

```python
ev_sales = ev[ev["parameter"] == "EV sales"]
ev_sales
```

```python
ev_sales_grouped = ev_sales.groupby("year")["value"].sum()
ev_sales_grouped
ev_sales_grouped.plot(x="year",y="value",kind="bar")
plt.xlabel("year")
plt.ylabel("EV Sales")
plt.show()
```

```python
ev_stock_grouped = ev_stock.groupby("year")["value"].sum().reset_index()
ev_sales_grouped = ev_sales.groupby("year")["value"].sum().reset_index()
merged_df = pd.merge(ev_stock_grouped,ev_sales_grouped,on = "year")
bar_width = 0.35
r1 = range(len(merged_df["year"]))
r2 = [x + bar_width for x in r1]
plt.figure(figsize=(10,6))
plt.bar(r1, merged_df['value_x'], color='b', width=bar_width,label = 'EV Stock')
plt.bar(r2, merged_df['value_y'], color='r', width=bar_width, label='EV Sales')
plt.xlabel('Year')
plt.xticks([r + bar_width/2 for r in range(len(merged_df['year']))], merged_df['year'],rotation = 45)
plt.ylabel('Number of Vehicles')
plt.title('EV Stocks and Sales per Year')
plt.legend()
plt.show()
```

```python
ev_electricity = ev[ev["parameter"] == "Electricity demand"]
ev_electricity
```

```python
ev_electricity.plot(x="year",y ="value",kind="line",color="b",label = "Electricity")
plt.xlabel("year")
plt.ylabel("GigaWatt Hour")
plt.title("Electricity Usage in GWH per year")
plt.show()
```
